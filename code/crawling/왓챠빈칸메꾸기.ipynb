```python
problem = HM[HM.제목_x.apply(lambda x: x.replace(" ", "") if type(x)==str else "1") != HM.제목_y.apply(lambda x: x.replace(" ", "") if type(x)==str else "")]
```

```python
problem.기타정보 = problem.기타정보.apply(lambda x: x[2:-2] if type(x)!=float else "").apply(lambda x: x.split(' ・ '))
```

```python
genre_list = ['드라마', '액션', '코미디', '로맨스', '다큐멘터리', '애니메이션', '공포', '스릴러', '판타지',
       '범죄', '로맨틱코미디', '전쟁', '모험',  'SF',  '가족', '스포츠',
       '미스터리', '음악', '서부극', '재난', '단편', '시대극', '성인', '인도', '역사','오스트리아', '틴에이저', '공연실황', '키즈']
```

```python
장르_y = []
연도_y = []
국가_y = []

for i in problem.index:
    if len(problem.loc[i, '기타정보']) >2:
        year = problem.loc[i, '기타정보'][0]
        
        if problem.loc[i, '기타정보'][1] in genre_list:
            genre = problem.loc[i, '기타정보'][1]
            country = problem.loc[i, '기타정보'][2:]
        else:
            genre = ''
            country = problem.loc[i, '기타정보'][1:]
        
    elif len(problem.loc[i, '기타정보']) ==2:
        year = problem.loc[i, '기타정보'][0]
        
        if problem.loc[i, '기타정보'][1] in genre_list:
            genre = problem.loc[i, '기타정보'][1]
            country = ''
        else:
            genre = ''
            country = problem.loc[i, '기타정보'][1]
        
    else:
        year = ''
        genre = ''
        country = ''
    
    연도_y.append(year)
    장르_y.append(genre)
    국가_y.append(country)
```

```python
problem.loc[:, '연도_y'] = 연도_y
problem.loc[:, '장르_y'] = 장르_y
problem.loc[:, '국가_y'] = 국가_y
```

```python
problem['연도'] = problem.연도.apply(lambda x: int(x) if x>2 else "").apply(str)
```

```python
missing_dict ={}
for i in list(problem[problem.제목_y_합 != ''].index):
    missing_dict[problem.loc[i, '제목_y_합']] = {'연도_y': problem.loc[i, '연도_y'],
                                                '장르_y':problem.loc[i, '장르_y'] ,
                                               '국가_y' :problem.loc[i, '국가_y'] ,
                                                 '별점' :problem.loc[i, '별점'] ,
                                                 '사람' :problem.loc[i, '사람'] ,
                                              '기타정보' : problem.loc[i, '기타정보']}
```
